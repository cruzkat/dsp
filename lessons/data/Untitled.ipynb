```python
import csv
from collections import defaultdict

def get_dict():
    faculty_dict = defaultdict(list)
    open_file = open('faculty.csv')
    read_file = csv.reader(open_file)
    faculty = list(read_file)
    faculty = faculty[1:]
    for row in faculty:
        name = row[0]
        name = name.split()
        last_name = name[-1]
        info = list(row[1:])
        if last_name in faculty_dict:
            faculty_dict[last_name].append(info)
        else:
            faculty_dict[last_name] = info
    return faculty_dict
```

```python
from csv import reader

def count_degrees(csv_file_name):
    open_file = open(csv_file_name)
    read_file = reader(open_file)
    faculty_csv = list(read_file)
    faculty_csv = faculty_csv[1:]
    degree_count = {}
    degree_list = []
    for row in faculty_csv:
        degree = row[1]
        degree = degree.replace("  ", " ").replace(".", '').upper()
        degree = degree.split()
        degree_list.extend(degree)
        
    for each in degree_list:
        if each in degree_count:
            degree_count[each] += 1
        else:
            degree_count[each] = 1
    return degree_count

count_degrees('faculty.csv')
```
